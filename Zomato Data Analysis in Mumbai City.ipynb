```python
# Values Counts in cuisines
zomato.Cuisines.value_counts()
```

```python
#let focus on a particular city and find that which kind of cusine is preferable

zomato_city = zomato[(zomato.City == 'Mumbai')]
zomato_city.head()
```

```python
# Value counts in city Cuisines column
zomato_city.Cuisines.value_counts()
```

```python
#top 15 popular cuisines
plt.figure(figsize=(10,6))
plot1 = sns.barplot(x = zomato_city.Cuisines.value_counts()[:15,].index, y = zomato_city.Cuisines.value_counts()[:15,], data = zomato_city, palette='rainbow')
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
plt.show()
```

```python
# Value counts in city locality column
zomato_city.Locality.value_counts()
```

```python
#top 15 popular cuisines
plt.figure(figsize=(10,6))
plot1 = sns.barplot(x = zomato_city.Locality.value_counts()[:15,].index, y = zomato_city.Locality.value_counts()[:15,], data = zomato_city, palette='rainbow')
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
plt.show()
```

```python
# let's focus on a particular area/locality : 
zomato_area = zomato_city[zomato_city.Locality == 'Malad West']
zomato_area.head()
```

```python
# Barplot of top cuisines
plt.figure(figsize=(12,10))
sns.barplot(x = 'Votes', y = 'Cuisines', data = zomato_area, palette='rainbow')
plt.show()
```

```python
# Bar plot as per votes to Restaurant
plt.figure(figsize=(12,15))
plot1 = sns.barplot(x = 'Votes', y = 'Restaurant_Name', data = zomato_area, palette='rainbow')
plt.show()
```

```python
# Bar plot as Aggregate rating to Cuisines
plt.figure(figsize=(12,10))
plot1 = sns.barplot(x = 'Aggregate rating', y = 'Cuisines', data = zomato_area, palette='rainbow')
plt.show()
```

```python
# Bar plot as Average_cost_for_two to Cuisines
plt.figure(figsize=(12,10))
plot1 = sns.barplot(x = 'Average_Cost_for_two', y = 'Cuisines', data = zomato_area, palette='rainbow')
plt.show()
```
